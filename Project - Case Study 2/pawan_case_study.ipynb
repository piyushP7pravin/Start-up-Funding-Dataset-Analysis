```python
d={}
si=df_copy[['StartupName','InvestorsName']]
for i in si.values:
    if str(i[1])!='nan' and str(i[1])!='Undisclosed Investors':
        inv_names=str(i[1]).split(',')
        for j in inv_names:
            
            if j.strip() in d:
                if i[0] not in d[j.strip()]:
                    d[j.strip()]=d[j.strip()]+[i[0]]
            else:
                
                d[j.strip()]=[i[0]]
for i in d:
    print(i,":",d[i])
    print()
```
